```python
# write your code here
import pandas as pd
import matplotlib.pyplot as plt  

df = pd.read_csv(r"C:\Users\nisri\OneDrive\Bureau\M111-Advanced-Python-and-Database-Management-main\day01\data\company_sales_data.csv")
month_number = df ['month_number'].tolist()
face_Cream = df ['facecream'].tolist()
face_Wash = df ['facewash'].tolist()
tooth_Paste = df ['toothpaste'].tolist()
bathing_soap = df ['bathingsoap'].tolist()
shampoo = df ['shampoo'].tolist()
moisturizer = df ['moisturizer'].tolist()

plt.plot(month_number, face_Cream,  label = 'Face cream Sales Data', marker='o', linewidth=3)
plt.plot(month_number, face_Wash,   label = 'Face Wash Sales Data',  marker='o', linewidth=3)
plt.plot(month_number, tooth_Paste, label = 'Tooth Paste Sales Data', marker='o', linewidth=3)
plt.plot(month_number, bathing_soap, label = 'Bathing soap Sales Data', marker='o', linewidth=3)
plt.plot(month_number, shampoo, label = 'Shampoo Sales Data', marker='o', linewidth=3)
plt.plot(month_number, moisturizer, label = 'Moisturizer Sales Data', marker='o', linewidth=3)

plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.legend(loc='upper left')
plt.yticks([1000, 2000, 4000, 6000, 8000, 10000, 12000, 15000, 18000])
plt.title('Sales data')
plt.show()
```
